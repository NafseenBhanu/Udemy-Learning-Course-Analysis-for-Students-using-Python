# Visualizations

### 1.Courses Analysis

```python
paid_perc=df['is_paid'].value_counts(normalize=True)
print(paid_perc)

plt.figure(figsize=(5,4))
plt.title('paid courses count')
plt.bar(['paid','not paid'],df['is_paid'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

### " most of the courses are paid courses.only few are free "

```python
sns.countplot(data=df,x='subject',hue='is_paid')
```

### " High concentration of web development and business finance courses "
### " Least amount courses are under the Graphic design "

### 2.Subject Analysis

```python
plt.figure(figsize=(14,7))
sns.histplot(data=df,x="subject",hue="subject")
plt.ylabel("num_subscribers")
plt.title("No. of Subscribers for each Subject")
plt.show()
```

### " Business Finance and web Development subjects have more number of subscribers "

### " users have a high subject preference towards web Development and Business Finance "

```python
plt.figure(figsize=(10,10))
sns.barplot(y="level",x="num_subscribers",data=df,width=0.8,orient="h",edgecolor="black")
plt.title("level vs num_subscribers")
plt.show()
```

### " Low concentration of Expert Level courses compared to the beginner and all level courses "
### " users have preferance over All level and Beginner category "

### 3.Correlation Matrix

```python
plt.figure(figsize=(17,6))
numeric_columns=df.select_dtypes(include="number")
plt.title("Correlation matrix")
sns.heatmap(numeric_columns.corr(),annot=True)
```

### 4.Content duration Analysis

```python
plt.figure(figsize=(14,8))
sns.scatterplot(x="level",y="content_duration",data=df,hue="level")
plt.title("level vs content_duration")
plt.show()
```

### " The content Duration is high for beginner and all level category  "

```python
plt.figure(figsize=(14,8))
sns.scatterplot(x="content_duration",y="num_subscribers",data=df)
plt.title("content_duration vs subscribers")
plt.show()
```

### " Most users prefer the courses to have less duration "

```python
plt.figure(figsize=(14,8))
sns.scatterplot(x="subject",y="num_reviews",data=df,hue="subject")
plt.title("subject vs reviews")
plt.show()
```

### " Web Development subject has more number of reviews "

```python
plt.figure(figsize=(15,9))
sns.boxplot(x="level",y="price",data=df)
plt.title("Distribution of price with level")
plt.show()
```

### " The price of the courses varied from 0 to 200 "
### " Expert level has the most concentration of high paid course "
### " Starting price of the beginner level courses are least "

### 5.Time series Analysis
